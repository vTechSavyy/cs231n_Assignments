```python
# Import libraries: 
import numpy as np
import time
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Generate the 2D data: 
N= 100  # No of points per class: 
D = 2  # Dimension of the data
      # Number of training examples
K = 3 # Number of classes

num_train = N*K

# Generate the training examples: 
# X - Input Training data - Matrix of dimension num_train x D 
# y - Vector of labels for each example - Vector of dimension num_train x 1
X = np.zeros((num_train,D))   # A tuple is needed for the dimensions only if the dimensions are greater than one
y = np.zeros(num_train, dtype='uint8')   # Is the default float?


# What ground truth function to use for the data??? Hmmm... 
# The spiral data shown in the CS231n toy example: 

for j in range(K):
  ix = range(N*j,N*(j+1))
  r = np.linspace(0.0,1,N) # radius
  t = np.linspace(j*4,(j+1)*4,N) + np.random.randn(N)*0.2 # theta
  X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
  y[ix] = j


# Visualize the generated data:Looks cool: 
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.Spectral)
plt.show()

# The important point to note is that the data is not linearly separable!!!
```

```python
# Task #1: Train a Linear softmax classifier on the data: 


# Steps: 
# - Split the data into training, validation and test datasets
# - Data pre-processing 
# - Setup the linear score function and cross-entropy loss function (I should understand what cross-entropy is)
    # Also add to it the regularization loss
# - Initialize the weights
# - Compute gradients analytically using back propagation
# - Check gradients numerically at certain examples to verify analytic formula
# - Allow the classifier to train its weights on the training data set. 
# - Tune hyperparameters using the validation dataset
# - Evaluate prediction accuracy on the test dataset. 


# Pre-define the softmax function: 
def softmax(x): 
    
    # Dimension of x is: num_train x num_classes
    
    # To ensure numeric stability: 
    x -= np.max(x)
    
    exp_x = np.exp(x)  
    
    return exp_x/(np.sum(np.exp(x), axis=1, keepdims= True))  # To understand this:::



# 1: Data split: 70 % training, 20 % testing , 10% validation: 

# X_train = X[:int(0.7*num_ex),:]
# X_test = X[int(0.7*num_ex):int(0.9*num_ex),:]
# y_test = y[int(0.7*num_ex):int(0.9*num_ex)]
# X_val = X[int(0.9*num_ex):,:]


# num_train = int(0.7*num_ex)

# print (X_train.shape)
# print (X_test.shape)
# print (X_val.shape)


# 2. Score and Loss functions: 

def compute_loss(X_train, W,b, reg):
    
    num_train = X_train.shape[0]

    # Dimension of lin_scores is - num_train x num_classes
    lin_scores = X_train.dot(W) + b


    # The loss for the ith training example is : The ratio of the exp( Score of the correct class )  to the ( Sum of exp (score of incorrect classes)))

    # Dimension of softmax scores is also - num_train x num_classes: 
    softmax_scores = softmax(lin_scores)

    cross_ent_loss = np.sum(-np.log(softmax_scores[np.arange(num_train), y[:num_train]]))/(num_train)


    #print ("The cross entropy loss for this set of weights is: ", cross_ent_loss)

    # Compute the regularization loss: 
    reg_loss = 0.5*reg*np.sum(W*W)
    
    
    loss = cross_ent_loss + reg_loss
    
    return loss , softmax_scores

    #print ("The regularization loss for this set of weights is: ", reg_loss)

```

```python
# Computing the gradient using Backpropagation: 

# The cross_entropy loss for each example is: L_i = -np.log(p_yi)

# Where, yi - Label for the ith example
#        p_k - Softmax function computed for the kth classes' score

# Start with the gradient of the cross_entropy_loss for each example w.r.t the weights: dL_i/dW

# Using chain rule split it into: dL_i/df_k * df_k/dW

# After getting the gradient for each example add them up to get the overall gradient for all examples.

def compute_grad(X_train, y, softmax_scores,reg):
    
    num_train = X_train.shape[0]
    
    # 1. Gradient of c.e_loss with respect to the scores: dL_i/df_k - Derived it in CS231n
    dscores = softmax_scores

    # For the correct class we subtract 1 from the gradient as proved in the derivation: 
    dscores[np.arange(num_train),y[:num_train]] -= 1

    # Divide by the num of training examples to account for the averaging of the loss: 
    dscores /= num_train

    # 2. Gradient of the scores w.r.t the weights: df_k/dW = X_i

    dW = np.dot(X_train.T,dscores)  # (num_dim x num_train) * (num_train x num_classes)

    # Gradient of the bias weights: dL_i/df_k itself except that we sum it over the training examples: 
    db = np.sum(dscores, axis=0, keepdims=True)

    # 3. Graident due to regularization loss: 
    dW += reg*W
    
    
    return dW,db


# print ("The overall gradient for the weights is: " , W)

# print (" The gradient of the c.e. loss w.r.t to the score for 1st 5 examples is: ", dscores[:5,:])

```

```python
# PUtting everything together for the training phase: 

# initialize parameters randomly
W = 0.01 * np.random.randn(D,K)
b = np.zeros((1,K))

# some hyperparameters
step_size = 1e-0
reg = 1e-3 # regularization strength
nIter = 200

for i in range(nIter): 
    
    loss, softmax_scores = compute_loss(X,W,b,reg)
    
    dW, db = compute_grad(X, y, softmax_scores,reg)
    
    W += -step_size*dW
    b += -step_size*db
    
    if i % 10 == 0:
        print ("iteration %d: loss %f" % (i, loss))
    


```

```python
# Prediction stage: Using the trained weights to evaluate accuracy of the test data set: 
pred_scores = X.dot(W) + b 

y_pred = np.argmax(pred_scores, axis =1)

print (" The accuracy of the Linear Softmax classifier is: ", np.mean(y_pred == y))
```

```python
### Non-linear classification using a neural network: 

## Step 1:  Setup the loss function: 

# Q: What is the loss function of the neural network? 
# Loss for ith example is: softmax(scores)-> Get softmax_score of correct class: 
# So here we are computing the probability of each class and then using the negative natural logarithm 
# of the correct class probability as the lossfor the ith example. The total loss is the mean loss for all the examples

def NN_loss(X,y,W1,b1,W2, b2,reg):
    
    num_train = X.shape[0]
    
    # Perform the forward pass of the network but break down the functions: 
    # Non-linearity - Implemented using a ReLU activation unit: 

    # 1st layer pass:
    # Dimensions of output are: num_train x num_hidden
    hidden_ac = X.dot(W1) + b1  # Hidden layer activations for all examples: 

    # Hidden layer non-linearity: ReLu unit: 
    hidden = np.maximum(0,hidden_ac)

    # 2nd layer pass: 
    # Dimensions of output are: num_train x num_classes: 
    scores = hidden.dot(W2)+ b2

    # First convert scores to softmax scores: 
    softmax_scores = softmax(scores)

    # Now compute the data loss: 
    data_loss = np.sum(-np.log(softmax_scores[np.arange(num_train),y]))/(num_train)

    # Now compute the regularization loss: 
    reg_loss = 0.5*reg*np.sum(W1*W1) + 0.5*reg*np.sum(W2*W2)  # Reg loss does not apply to the bias weights: 
    
    loss = data_loss + reg_loss
    
    return loss, softmax_scores, hidden

def NN_grad(X,y,softmax_scores, hidden, W1,b1,W2, b2,reg):
    
    num_train = X.shape[0]
    
    ## Step 2: Apply backpropagation to compute the gradient of the loss function w.r.t the weights: 

    # 2.a: Compute the gradient at the output layer: This is the gradient of the loss function w.r.t the raw scores: 
    # Similar to the linear softmax classifier derivation this can be computed as: 
    dscores = softmax_scores # Except we need to subtract 1 from the probability of the correct class. 

    dscores[np.arange(num_train),y] -= 1
    
    dscores /= num_train

    # 2.b Compute the gradient at the second layer: i.e gradient of loss w.r.t W2 and b2: 
    # Now L_i = softmax(W2*h +b2): 

    dW2 = np.dot(hidden.T,dscores)   # The mathematical expression : dW2 = h^T*dscores  (num_hidden x numtrain ) * (num_train x num_classes)
    db2 = np.sum(dscores, axis=0, keepdims = True)

    # 2.c : Compute the gradient at the first layer: i.e gradient of loss w.r.t W1 and b1: 
    # Dimensions: dhidden = num_train x num_hidden: 
    # dscores = num_train x num_classes
    # W2 = num_hidden x num_classes
    dhidden = np.dot(dscores, W2.T)

    # Apply the ReLU activation unit gradient: 
    # Dimension: dhidden - num_train x num_hidden
    dhidden[hidden<=0] = 0

    # Compute the gradient for weights in the first layer: 
    # Dimension - dW1 - num_dim x num_hidden
    # db1 - 1 x num_hidden
    # dhidden = num_train x num_hidden
    # X - num_train x num_dim
    dW1 = np.dot(X.T, dhidden)
    db1 = np.sum(dhidden, axis=0, keepdims = True)
    
    dW1 += reg*W1
    dW2 += reg*W2
    
    return dW1, db1, dW2, db2



# How to perform gradient checks now: 
```

```python
## Step 3: Training the neural network: 
nIter = 10000
step_size_NN = 0.1
reg = 1e-3 # regularization strength

# TBD: Number of hidden layers and number of neurons in the hidden layer: 
# Number of hidden layers chosen is 1
h = 100  # Why did Andrej choose such a high number of neurons in the hidden layer

# Initialize the weights for each layer: 
# Layer #1
W1 = 0.01*np.random.rand(D,h)
b1 = np.zeros((1,h))

# Layer #2: 
W2 = 0.01*np.random.rand(h,K)
b2 = np.zeros((1,K))

# gradient descent loop
num_examples = X.shape[0]

for i in range(nIter): 
    
    # Compute the loss for the current iteration using all training samples: 
    loss, probs, hidden = NN_loss(X,y,W1,b1,W2,b2,reg)
    
#     #evaluate class scores, [N x K]
#     #X - num_train x num_dim
#     #W1 - num_dim x num_hidden
#     #hidden_layer - num_train x num_hidden
#     hidden_layer = np.maximum(0, np.dot(X, W1) + b1) # note, ReLU activation
    
#     # W2 - num_hidden x num_classes
#     # scores - num_train x num_classes
#     scores = np.dot(hidden_layer, W2) + b2

#     # compute the class probabilities
#     exp_scores = np.exp(scores)
#     probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True) # [N x K]

#     # compute the loss: average cross-entropy loss and regularization
#     corect_logprobs = -np.log(probs[range(num_examples),y])
#     data_loss = np.sum(corect_logprobs)/num_examples
#     reg_loss = 0.5*reg*np.sum(W1*W1) + 0.5*reg*np.sum(W2*W2)
#     loss = data_loss + reg_loss

    # Compute the gradient of the loss w.r.t to the weights: 
    dW1 , db1 , dW2, db2 = NN_grad(X,y,probs,hidden, W1,b1,W2,b2,reg)
    
#     #compute the gradient on scores
#     dscores = probs
#     dscores[range(num_examples),y] -= 1
#     dscores /= num_examples

#     # backpropate the gradient to the parameters
#     # first backprop into parameters W2 and b2
#     dW2 = np.dot(hidden_layer.T, dscores)
#     db2 = np.sum(dscores, axis=0, keepdims=True)
#     # next backprop into hidden layer
#     dhidden = np.dot(dscores, W2.T)
#     # backprop the ReLU non-linearity
#     dhidden[hidden_layer <= 0] = 0
#     # finally into W,b
#     dW1 = np.dot(X.T, dhidden)
#     db1 = np.sum(dhidden, axis=0, keepdims=True)

#     # add regularization gradient contribution
#     dW2 += reg * W2
#     dW1 += reg * W1
    
    # Update the parameters: 
    W1 += -step_size_NN*dW1
    b1 += -step_size_NN*db1
    W2 += -step_size_NN*dW2
    b2 += -step_size_NN*db2
    
    
    # Print out the loss for the current iteration: 
    if i % 1000 == 0:
        print ("iteration %d: loss %f" % (i, loss))
        #print ("W1 gradient is: ", dW1[0:10,0:10])
    
```

```python
# initialize parameters randomly
h = 100 # size of hidden layer
W = 0.01 * np.random.randn(D,h)
b = np.zeros((1,h))
W2 = 0.01 * np.random.randn(h,K)
b2 = np.zeros((1,K))

# some hyperparameters
step_size = 1e-0
reg = 1e-3 # regularization strength

# gradient descent loop
num_examples = X.shape[0]
for i in range(10000):
  
  # evaluate class scores, [N x K]
  hidden_layer = np.maximum(0, np.dot(X, W) + b) # note, ReLU activation
  scores = np.dot(hidden_layer, W2) + b2
  
  # compute the class probabilities
  exp_scores = np.exp(scores)
  probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True) # [N x K]
  
  # compute the loss: average cross-entropy loss and regularization
  corect_logprobs = -np.log(probs[range(num_examples),y])
  data_loss = np.sum(corect_logprobs)/num_examples
  reg_loss = 0.5*reg*np.sum(W*W) + 0.5*reg*np.sum(W2*W2)
  loss = data_loss + reg_loss
  if i % 1000 == 0:
    print ("iteration %d: loss %f" % (i, loss))
  
  # compute the gradient on scores
  dscores = probs
  dscores[range(num_examples),y] -= 1
  dscores /= num_examples
  
  # backpropate the gradient to the parameters
  # first backprop into parameters W2 and b2
  dW2 = np.dot(hidden_layer.T, dscores)
  db2 = np.sum(dscores, axis=0, keepdims=True)
  # next backprop into hidden layer
  dhidden = np.dot(dscores, W2.T)
  # backprop the ReLU non-linearity
  dhidden[hidden_layer <= 0] = 0
  # finally into W,b
  dW = np.dot(X.T, dhidden)
  db = np.sum(dhidden, axis=0, keepdims=True)
  
  # add regularization gradient contribution
  dW2 += reg * W2
  dW += reg * W
  
  # perform a parameter update
  W += -step_size * dW
  b += -step_size * db
  W2 += -step_size * dW2
  b2 += -step_size * db2
```
